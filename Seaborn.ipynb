Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative
statistical graphics. Seaborn is built on top of matplotlib and integrates closely with pandas data structures. It helps you explore and
understand your data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.clustermap(iris.iloc[:, [0, 1, 2, 3]])
```

```python
# scatterplot in category
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
# figure level
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill',
hue='smoker', estimator=np.min)
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist', hue='sex')
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.boxplot)
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.kdeplot, sns.histplot)
```

```python
g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
